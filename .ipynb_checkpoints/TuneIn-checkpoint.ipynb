```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="my-application")

tuneindata['BusinessLocation'] = [geolocator.geocode(x, timeout = 60) for x in tuneindata['DBAName']]
x = tuneindata['BusinessLocation']
tuneindata.dropna(subset=['BusinessLocation'], inplace=True)
tuneindata['g_latitude'] = tuneindata['BusinessLocation'].apply(lambda loc: loc.latitude)
tuneindata['g_longitude'] = tuneindata['BusinessLocation'].apply(lambda loc: loc.longitude)
tuneindata
```
